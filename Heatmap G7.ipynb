```python
import folium
```

```python
from pyproj import crs
import pandas as pd
import matplotlib.pyplot as plt
from folium.plugins import HeatMap
import numpy as np
from folium.plugins import MarkerCluster
```

```python
# Create a Map instance
m = folium.Map(location=[48.8639368381152, 2.349781581911858], zoom_start=6, control_scale=True)
```

```python
locations2 = pd.read_excel('20220509_heatmap G7_locations.xlsx')
```

```python
# Following this example: https://github.com/python-visualization/folium/blob/master/examples/MarkerCluster.ipynb


# Create a list of coordinate pairs
locations_list2 = list(zip(locations2["Lattitude"], locations2["Longitude"]))
```

```python
# Create a Map instance
m = folium.Map(location=[48.8639368381152, 2.349781581911858], tiles = 'cartodbpositron', zoom_start=6, control_scale=True)
```

```python
# Create a folium marker cluster
marker_cluster = MarkerCluster(locations_list2)

# Add marker cluster to map
marker_cluster.add_to(m)

# Show map
m
```

```python
outfp = "heatmap_g7.html"
m.save(outfp)
```
